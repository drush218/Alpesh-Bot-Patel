```python
import pandas as pd


def pct_change(old,new):
    return (new - old) / old


```

```python
#Input current price
CurrentPrice = 42

#Input price target and calculate upside
PriceTarget = 52
ImpliedUpside = pct_change(CurrentPrice, PriceTarget)

print("Implied Upside =", ImpliedUpside)

#Input stop lose and calculate downside
StopLoss = 39
ImpliedDownside = abs(pct_change(CurrentPrice, StopLoss))

print("Implied Downside =", ImpliedDownside)
```

```python
#Calculate reward to risk ratio

R2R = ImpliedUpside/ImpliedDownside

print("Reward:Risk =", R2R)
```

```python
import pandas as pd

# Import portfolio value
PortfolioValue = 12500

# Set max acceptable loss as percentage of portfolio

MaxLoss_pct = 0.01
#Hence
MaxLoss = MaxLoss_pct * PortfolioValue

# Set MACD signals
MACD_Monthly = 0
MACD_Weekly = 0

# Compute MACD Score
MACD_Score = MACD_Weekly + MACD_Monthly

# Define MACD matrix
MACD_Matrix = pd.DataFrame({
    'MACD Score': [0, 1, 2],
    'pct loss' : [2, 1.5, 1]   # percent as number, e.g., 0.5%
})

# Lookup downside risk for the score
DownsideRiskFactor = MACD_Matrix.loc[MACD_Matrix['MACD Score'] == MACD_Score, 'pct loss'].iloc[0]

# Consider upside if MACD score is high
UpsideRewardFactor = R2R / 2 if DownsideRiskFactor == 1 else 1

AdjustmentFactor = UpsideRewardFactor / DownsideRiskFactor

# Calculate investment value based on stop loss and maximum acceptable loss
InvestmentValue = MaxLoss / ImpliedDownside


# Adjust total investment for risk and reward
InvestmentValue_Adj = InvestmentValue * AdjustmentFactor
print("Suggested investment value =", InvestmentValue_Adj)
```
